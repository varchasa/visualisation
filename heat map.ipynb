```python
f=sns.load_dataset('flights')
```

```python
f.head()
```

```python
f.sample(10)
```

```python
f_ = f.pivot(index='year', columns='month', values='passengers');
```

```python
f_.head()
```

```python
sns.heatmap(f_.T)
```

```python
fig = plt.gcf();
fig.set_size_inches(15, 10)
sns.heatmap(f_.T, annot=True, fmt="d");
```

```python
fig = plt.gcf();
fig.set_size_inches(15, 10)
sns.heatmap(f_.T, annot=True, fmt="d", cmap="YlGnBu");
```

```python
fig = plt.gcf();
fig.set_size_inches(15, 10)
sns.heatmap(f_.T, annot=True, fmt="d", 
            cmap=sns.diverging_palette(50, 200, n=45));
```

```python
fig = plt.gcf();
fig.set_size_inches(15, 10)
sns.heatmap(f_.T, annot=True, fmt="d", 
            cmap=sns.diverging_palette(250, 10, n=45),
            center=f_.loc[1954, 'January']);
```
