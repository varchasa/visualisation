## Pie chart

```python
p=sns.load_dataset('penguins')
```

```python
p.head()
```

```python
p.groupby('species')['species'].count()
```

```python
c=p.groupby('species')['species'].count()
```

```python
plt.pie(c);
plt.show()
```

```python
plt.pie(c, labels=c.index);
plt.show()
```

```python
plt.pie(c, labels=c.index, autopct="%.2f%%");
plt.show()
```

```python
plt.pie(c, labels=c.index, autopct="%.2f%%",
        explode=[0, 0, 1]);
plt.show()
```

```python
plt.pie(c, labels=c.index, autopct="%.2f%%",
        explode=[0, 1, 0]);
plt.show()
```

```python
plt.pie(c, labels=c.index, autopct="%.2f%%",
        explode=[1, 0, 0]);
plt.show()
```

```python
plt.pie(c, labels=c.index, autopct="%.2f%%",
        explode=[1, 1, 0]);
plt.show()
```
