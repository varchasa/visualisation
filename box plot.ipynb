# Box plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x=np.random.normal(size=1000)
```

```python
sns.boxplot(x);
```

```python
x = np.random.uniform(size=1000)
```

```python
sns.boxplot(x);
```

```python
sns.boxplot(x, whis=0.2)
```

```python
x = np.random.normal(size=1000)
```

```python
sns.boxplot(x, whis=0.5,fliersize=1);
```

```python
sns.boxplot(x, whis=0.5, fliersize=1, orient='v');
```
